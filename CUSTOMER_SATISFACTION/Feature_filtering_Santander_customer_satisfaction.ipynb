```python
import pandas as pd
import numpy as np
import matplotlib as mp
```

### LOAD DATA

```python
df = pd.read_csv('train.csv')
```

```python
df
```

### Setting first 1500 rows to dataframe

```python
df = df.head(1500)
```

### Splitting the dataframe to train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test = train_test_split(df)
x_train
```

```python
### Removing constant features
* Identifying those features whose values do not vary much and removing them
* Constant features are the ones whose variance is less than 0.5
```

```python
column_names = df.columns
constant_features = []
for col in column_names:
    if x_train[col].var() <= 1:
        constant_features.append(col)
        del x_train[col]
        del x_test[col]
```

```python
x_train
```

```python
print("Total no. of constant features:", len(constant_features))
for i in range(1,len(constant_features)+1):
    print(i,constant_features[i-1])
```

```python
### Removing duplicate features
* Duplicate features are those whose data are same as another feature of data frame.
* Duplicate feature might have a different feature name but same values.
```

```python
import itertools
combos = list(itertools.combinations(x_train.columns,2))
duplicate_rows = []
for f1,f2 in combos:
    if f1 not in duplicate_rows and f2 not in duplicate_rows:
        if x_train[f1].equals(x_train[f2]):
            duplicate_rows.append(f1)
            del x_train[f1]
```

```python
print("Total no. of constant features:", len(duplicate_rows))
for i in range(1,len(duplicate_rows)+1):
    print(i,duplicate_rows[i-1])
```

```python
### Final train data
```

```python
print("No. of feature columns removed = ", len(constant_features)+len(duplicate_rows))
```
