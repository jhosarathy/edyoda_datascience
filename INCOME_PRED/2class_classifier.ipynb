```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cols = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship'
        ,'race','sex','capital-gain','capital-loss','hours-per-week','native-country','Salary']
adult_data = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/adult.data.txt', names=cols)
```

```python
df = adult_data 
```

### Addressing "Take care of missing values marked by '?'"

```python
for item in cols:
    if(df[item].dtypes == df['Salary'].dtypes):
        df.drop(df[df[item].str.match('\s*\?\s*')].index, inplace=True)
```

```python
df.tail()
```

### No of records
* No of records present
* The number of individuals making more or less than 50k etc, will show us how many individuals fit in each group.

```python
n_records = df.shape[0]
n_greater_50k = df[df['Salary'] == ' >50K'].shape[0]
n_atmost_50k = df[df['Salary'] == ' <=50K'].shape[0]
print("No.of records present:", n_records)
print("No.of individuals making more than 50k:", n_greater_50k)
print("No.of individuals making less than or equal to 50k:", n_atmost_50k)
```

```python
from sklearn.preprocessing import MinMaxScaler
df_tran = pd.DataFrame(data = df)
numerical = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
df_tran[numerical] = MinMaxScaler().fit_transform(df[numerical])
```

```python
objects = []
for col in cols:
    if df[col].dtypes == df['Salary'].dtypes:
        objects.append(col)
from sklearn.preprocessing import LabelEncoder
for i in objects:
    df_tran[i] = LabelEncoder().fit_transform(df[i])
```

```python
features = df_tran.iloc[:, :-1]
```

```python
from sklearn.model_selection import train_test_split
trainX, testX, trainY, testY = train_test_split(features, target, train_size = 0.3 )
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnc = KNeighborsClassifier()
knnc.fit(trainX,trainY)
knnc.score(testX, testY)
```
