```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cols = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship'
        ,'race','sex','capital-gain','capital-loss','hours-per-week','native-country','Salary']
adult_data = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/adult.data.txt', names=cols)
```

```python
df = adult_data 
```

### Addressing "Take care of missing values marked by '?'"

```python
for item in cols:
    if(df[item].dtypes == df['Salary'].dtypes):
        df.drop(df[df[item].str.match('\s*\?\s*')].index, inplace=True)
```

```python
df
```

### No of records
* No of records present
* The number of individuals making more or less than 50k etc, will show us how many individuals fit in each group.

```python
n_records = df.shape[0]
n_greater_50k = df[df['Salary'] == ' >50K'].shape[0]
n_atmost_50k = df[df['Salary'] == ' <=50K'].shape[0]
print("No.of records present:", n_records)
print("No.of individuals making more than 50k:", n_greater_50k)
print("No.of individuals making less than or equal to 50k:", n_atmost_50k)
```
